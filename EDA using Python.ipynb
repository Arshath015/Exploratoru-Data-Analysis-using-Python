```python
df = pd.read_csv(r'D:\Projects\EDA\StudentsPerformance.csv')
```

```python
df_numeric = df.select_dtypes(include=[float, int])
```

```python
correlation = df_numeric.corr()
print(correlation)
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,annot=True) 
```

```python
sns.pairplot(students)
```

```python
sns.relplot(x='math score', y='reading score',hue='gender',data=students)
sns.relplot(x='math score', y='writing score',hue='gender',data=students)
sns.relplot(x='reading score', y='writing score',hue='gender',data=students)
#just like that you can make relation plot for all the combinations of variables to visualize relationships between two variables and can also include a third variable through color coding. 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
#sns.distplot(students['math score'])
#sns.distplot(students['reading score'])
#sns.distplot(students['writting score'])
sns.histplot(df['math score'], kde=True, color='blue', label='math score', bins=10)
sns.histplot(df['reading score'], kde=True, color='green', label='Reading Score', bins=10)
sns.histplot(df['writing score'], kde=True, color='red', label='Writing Score', bins=10)

plt.legend()
```
